```python
# import the necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# import the data
data = pd.read_csv("/content/apple_quality.csv")
```

## Data Visualization

```python
# Data visualization
sns.pairplot(data)
plt.show()
```

```python
# Creating the correlation heatmap
corr = data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Distribution the size of Apples
plt.figure(figsize=(10, 6))
sns.histplot(data['Size'], bins=20, kde=True, color='orange', label='Size')
plt.title('Distribution of Size')
plt.xlabel('Size')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

```python
features=['Weight','Sweetness','Crunchiness','Juiciness','Ripeness','Acidity']
for i in features :
  plt.scatter(data[i], data['Size'])
  plt.xlabel(i)
  plt.ylabel('Size')
  plt.title(f'Scatter Plot: Size vs {i}')
  plt.show()
```
